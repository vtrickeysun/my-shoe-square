```python
%matplotlib inline
```

```python
df1 = df[df['Edition'] == 1896]
df1.head()
```

# Line Plot

```python
df1['Sport'].value_counts().plot(kind = 'line')
plt.xticks(rotation=45)
plt.show()
```

# Bar plot

```python
df1['Sport'].value_counts().plot(kind = 'bar');
plt.xticks(rotation=45)
plt.show()
```

### Horizontal bar plot

```python
df1['Sport'].value_counts().plot(kind = 'barh');
plt.yticks(rotation=45)
plt.show()
```

### Pie chartm

```python
df1['Sport'].value_counts().plot(kind = 'pie');
```
